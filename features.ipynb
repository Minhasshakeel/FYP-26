# Pipeline

# Indoor Sports

```python
# Normalization Function
def normalize_dataset(df):
    df = df.copy()
    numerical_cols = df.select_dtypes(include=np.number).columns
    scaler = StandardScaler()
    df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
    return df
```

```python
def clean_columns(df):
    df.columns = df.columns.str.strip().str.lower()
    return df
```

# Cricket

```python
# def process_cricket(df):
    
#     new_df = pd.DataFrame()
    
#     new_df["Performance_Score"] = df["Runs"]
#     new_df["Offensive_Skill"] = df["Fours_x"] + df["Sixes_x"]
#     new_df["Defensive_Skill"] = df["Wicket Taken"]
#     new_df["Efficiency"] = df["Avg Bat Strike Rate"]
#     new_df["Aggression"] = df["Run Rate"]
#     new_df["Mistake_Penalty"] = df["Wicket Lost"]
#     new_df["Experience"] = 0
#     new_df["Physical_Index"] = 0
    
#     new_df["Game_Type"] = "Cricket"
    
#     return normalize_dataset(new_df)
def process_cricket(df):
    
    df = clean_columns(df)
    
    new_df = pd.DataFrame()
    new_df["performance_score"] = df["runs"]
    new_df["offensive_skill"] = df["fours_x"] + df["sixes_x"]
    new_df["defensive_skill"] = df["wicket taken"]
    new_df["efficiency"] = df["avg bat strike rate"]
    new_df["aggression"] = df["run rate"]
    new_df["mistake_penalty"] = df["wicket lost"]
    new_df["experience"] = 0
    new_df["physical_index"] = 0
    new_df["game_type"] = "Cricket"
    return normalize_dataset(new_df)
```

# Football

```python
# def process_football(df):
#     new_df = pd.DataFrame()

#     new_df["Performance_Score"] = df["overall"]
#     new_df["Offensive_Skill"] = df["shooting"] +["pace"] + ["dribbling"]
#     new_df["Defensive_Skill"] = df["defending"]
#     new_df["Efficiency"] = df["potential"]
#     new_df["Aggression"] = df["physic"]
#     new_df["Mistake_Penalty"] = 0
#     new_df["Experience"] = df["age"]
#     new_df["Physical_Index"] = df["height_cm"] + ["weight_cm"]
    
#     new_df["Game_Type"] = df["Football"]

#     return normalize_dataset(new_df)

def process_football(df):
    df = clean_columns(df)
    new_df = pd.DataFrame()
    new_df["performance_score"] = df["overall"]
    new_df["offensive_skill"] = df["shooting"] + df["pace"] + df["dribbling"]
    new_df["defensive_skill"] = df["defending"]
    new_df["efficiency"] = df["potential"]
    new_df["aggression"] = df["physic"]
    new_df["mistake_penalty"] = 0
    new_df["experience"] = df["age"]
    new_df["physical_index"] = df["height_cm"] + df["weight_kg"]
    new_df["game_type"] = "Football"
    return normalize_dataset(new_df)
```

# Golf

```python
# def process_golf(df):
#     new_df = pd.DataFrame()

#     new_df["Performance_Score"] = df["SCORE"]
#     new_df["Offensive_Skill"] = df["FED PTS"]
#     new_df["Defensive_Skill"] = df["TOT"]
#     new_df["Efficiency"] = df["TOT"]
#     new_df["Aggression"] = df["R1"] + df["R2"] + df["R3"] + df["R4"]
#     new_df["Mistake_Penalty"] = 0
#     new_df["Experience"] = df["EARNINGS"]
#     new_df["Physical_Index"] = 0
    
#     new_df["Game_Type"] = df["golf"]

#     return normalize_dataset(new_df)
def process_golf(df):
    df = clean_columns(df)
    new_df = pd.DataFrame()
    new_df["performance_score"] = df["score"]
    new_df["offensive_skill"] = df["r1"] + df["r2"] + df["r3"] + df["r4"]
    new_df["defensive_skill"] = df["tot"]
    new_df["efficiency"] = df["fedex pts"]
    new_df["aggression"] = df["earnings"]
    new_df["mistake_penalty"] = 0
    new_df["experience"] = 0
    new_df["physical_index"] = 0
    new_df["game_type"] = "Golf"
    return normalize_dataset(new_df)
```

# Rugby

```python
def process_rugby(df):
    df = clean_columns(df)
    
    # Fix duplicates if any
    df = df.loc[:, ~df.columns.duplicated()]
    
    new_df = pd.DataFrame()
    new_df["performance_score"] = df["pts"]
    new_df["offensive_skill"] = df["try"] + df["pen"]
    new_df["defensive_skill"] = df["national_w"]
    new_df["efficiency"] = df["national_w"]
    new_df["aggression"] = df["min"]
    
    # Mistake penalty (yellow + red), ensure columns exist
    yellow_col = "yellow" if "yellow" in df.columns else 0
    red_col = "red" if "red" in df.columns else 0
    new_df["mistake_penalty"] = df[yellow_col] + df[red_col]
    
    new_df["experience"] = df["played"] if "played" in df.columns else 0
    new_df["physical_index"] = df["weight"] + df["tall(m)"]
    new_df["game_type"] = "Rugby"
    
    return normalize_dataset(new_df)
```

# Badminton

```python
# def process_badminton(df):
    
#     new_df = pd.DataFrame()
    
#     new_df["Performance_Score"] = df["Points_player1"]
#     new_df["Offensive_Skill"] = df["player1_set1"] + df["player1_set2"] + df["player1_set3"]
#     new_df["Defensive_Skill"] = df["Points_player2"]
#     new_df["Efficiency"] = 1 / df["Rank_player1"]
#     new_df["Aggression"] = df["Points_player1"]
#     new_df["Mistake_Penalty"] = df["Points_player2"]
#     new_df["Experience"] = df["year"]
#     new_df["Physical_Index"] = 0  # not available
    
#     new_df["Game_Type"] = "Badminton"
    
#     return normalize_dataset(new_df)

def process_badminton(df):
    df = clean_columns(df)
    new_df = pd.DataFrame()
    new_df["performance_score"] = df["points_player1"]
    new_df["offensive_skill"] = df["player1_set1"] + df["player1_set2"] + df["player1_set3"]
    new_df["defensive_skill"] = df["points_player2"]
    new_df["efficiency"] = 1 / (df["rank_player1"] + 1)
    new_df["aggression"] = df["points_player1"]
    new_df["mistake_penalty"] = df["points_player2"]
    new_df["experience"] = df["year"]
    new_df["physical_index"] = 0
    new_df["game_type"] = "Badminton"
    return normalize_dataset(new_df)
```

# Baseball

```python
# def process_baseball(df):
    
#     new_df = pd.DataFrame()
    
#     new_df["Performance_Score"] = df["Runs"]
#     new_df["Offensive_Skill"] = df["HR"] + df["HmRun"]
#     new_df["Defensive_Skill"] = df["PutOuts"] + df["Assists"]
#     new_df["Efficiency"] = df["AVG"]
#     new_df["Aggression"] = df["AtBat"]
#     new_df["Mistake_Penalty"] = df["SO"]
#     new_df["Experience"] = 0
#     new_df["Physical_Index"] = 0
    
#     new_df["Game_Type"] = "Baseball"
    
#     return normalize_dataset(new_df)

def process_baseball(df):
    df = clean_columns(df)
    
    # Ensure numeric
    df["hr"] = pd.to_numeric(df["hr"], errors="coerce").fillna(0)
    df["hmrun"] = pd.to_numeric(df["hmrun"], errors="coerce").fillna(0)
    df["putouts"] = pd.to_numeric(df["putouts"], errors="coerce").fillna(0)
    df["assists"] = pd.to_numeric(df["assists"], errors="coerce").fillna(0)
    df["avg"] = pd.to_numeric(df["avg"], errors="coerce").fillna(0)
    df["atbat"] = pd.to_numeric(df["atbat"], errors="coerce").fillna(0)
    df["so"] = pd.to_numeric(df["so"], errors="coerce").fillna(0)
    
    new_df = pd.DataFrame()
    new_df["performance_score"] = df["runs"]
    new_df["offensive_skill"] = df["hr"] + df["hmrun"]
    new_df["defensive_skill"] = df["putouts"] + df["assists"]
    new_df["efficiency"] = df["avg"]
    new_df["aggression"] = df["atbat"]
    new_df["mistake_penalty"] = df["so"]
    new_df["experience"] = 0
    new_df["physical_index"] = 0
    new_df["game_type"] = "Baseball"
    
    return normalize_dataset(new_df)
```

# Table Tennis

```python
# def process_table_tennis(df):
    
#     new_df = pd.DataFrame()
    
#     new_df["Performance_Score"] = df["Sets_P1"]
#     new_df["Offensive_Skill"] = df["P1_G1"] + df["P1_G2"]
#     new_df["Defensive_Skill"] = df["Sets_P2"]
#     new_df["Efficiency"] = df["HomeWinner"]
#     new_df["Aggression"] = df["Sets_P1"]
#     new_df["Mistake_Penalty"] = df["Sets_P2"]
#     new_df["Experience"] = 0
#     new_df["Physical_Index"] = 0
    
#     new_df["Game_Type"] = "Table_Tennis"
    
#     return normalize_dataset(new_df)

def process_table_tennis(df):
    df = clean_columns(df)
    new_df = pd.DataFrame()
    new_df["performance_score"] = df["sets_p1"]
    new_df["offensive_skill"] = df["p1_g1"] + df["p1_g2"]
    new_df["defensive_skill"] = df["sets_p2"]
    new_df["efficiency"] = 1
    new_df["aggression"] = df["sets_p1"]
    new_df["mistake_penalty"] = df["sets_p2"]
    new_df["experience"] = 0
    new_df["physical_index"] = 0
    new_df["game_type"] = "TableTennis"
    return normalize_dataset(new_df)
```

# Volley Ball

```python
# def process_volleyball(df):
    
#     new_df = pd.DataFrame()
    
#     new_df["Performance_Score"] = df["attack"]
#     new_df["Offensive_Skill"] = df["attack"] + df["serve"]
#     new_df["Defensive_Skill"] = df["block"] + df["dig"] + df["receive"]
#     new_df["Efficiency"] = df["set"]
#     new_df["Aggression"] = df["attack"]
#     new_df["Mistake_Penalty"] = 0
#     new_df["Experience"] = df["age"]
#     new_df["Physical_Index"] = 0
    
#     new_df["Game_Type"] = "Volleyball"
    
#     return normalize_dataset(new_df)

def process_volleyball(df):
    df = clean_columns(df)
    new_df = pd.DataFrame()
    new_df["performance_score"] = df["attack"]
    new_df["offensive_skill"] = df["attack"] + df["serve"]
    new_df["defensive_skill"] = df["block"] + df["dig"] + df["receive"]
    new_df["efficiency"] = df["set"]
    new_df["aggression"] = df["attack"]
    new_df["mistake_penalty"] = df["block"]
    new_df["experience"] = 0
    new_df["physical_index"] = 0
    new_df["game_type"] = "Volleyball"
    return normalize_dataset(new_df)
```

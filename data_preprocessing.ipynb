```python
cric_df3 = pd.read_csv("Outdoor/Cricket/batsman data odi.csv")
cric_df3.head()
```

```python
cric_df3.isnull().sum()
```

```python
columns_to_fill = [
    'Country',
    'Time Period',
    'Matches',
    'Played',
    'Not Outs',
    'Runs',
    'Highest Score',
    'Batting Average',
    'Balls Faced',
    'Strike Rate',
    'Centuries',
    'Fifties',
    'Ducks',
    'Fours',
    'Sixes'
]

for col in columns_to_fill:
    if col in cric_df3.columns:
        cric_df3[col] = cric_df3[col].fillna(0)
```

```python
columns_to_drop = ['Country', 'Time Period', 'Matches', 'Ducks',
]
cric_df3.drop(columns=columns_to_drop, inplace=True)
```

```python
cric_df3.head()
```

```python
match_df.isna().sum()
```

```python
final_cric = match_df.merge(
    batsman_agg,
    on="Runs",
    how="left"
)
```
